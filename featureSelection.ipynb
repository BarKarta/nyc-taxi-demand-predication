```python
main_df=pd.read_csv(r'post_data_prep_all_zones.csv')

main_df.dropna(inplace=True)
main_df.drop(main_df[main_df['trip_time'] < 5].index,axis=0,inplace=True) 
weekday_df = pd.get_dummies(data=main_df.weekday)#, drop_first=True)
main_df = pd.merge(left=main_df, right=weekday_df, right_index=True, left_index=True)

time_binned_df = pd.get_dummies(data=main_df.time_binned)#, drop_first=True)
main_df = pd.merge(left=main_df, right=time_binned_df, right_index=True, left_index=True)

main_df.drop(['weekday','time_binned'],axis=1,inplace=True)
main_df.drop(['pickup_date','Tmax','Tmin','Tavg','Tdep','HDD','CDD','Precipitation','new_snow','snow_depth'],axis=1,inplace=True)


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(main_df[['trip_time','speed','trip_distance']]), columns=['trip_time_scaled','speed_scaled','distance_scaled'])

not_scaled_df=main_df.drop(['trip_distance'	,'trip_time'	,'speed'],axis=1)

scaled_main_df = pd.merge(left=scaled_df, right=not_scaled_df, left_index=True, right_index=True)

scaled_main_df



```

```python
from sklearn.model_selection import train_test_split
X=scaled_main_df.drop(['num_of_taxis'],axis=1)
y=scaled_main_df.num_of_taxis

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
train_df = pd.merge(left=X_train, right=y_train, left_index=True, right_index=True)
test_df = pd.merge(left=X_test, right=y_test, left_index=True, right_index=True)
test_df.head(10)


from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50,criterion='entropy')

rfc.fit(X_train,y_train)

```
